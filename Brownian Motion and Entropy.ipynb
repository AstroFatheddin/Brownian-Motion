```python
import matplotlib.pyplot as plt
import numpy as np
r=[]
m=10
for i in range(-m,m+1):
    for j in range(-m,m+1):
        r.append([i,j])
ra=np.array(r)
plt.xlim(-100,100)
plt.ylim(-100,100)
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal',adjustable='box')
plt.plot(ra[:,0],ra[:,1],'o')
plt.show()
```

```python
import random
m=10
N=(2*m+1)*(2*m+1)
L=100
Nstep=5000000
Ni=[]
Nj=0
S=[]
count=np.zeros((8,8))
for k in range(Nstep):
    n=random.randrange(N)
    x0=r[n][0]
    y0=r[n][1]
    p=random.randrange(4)
    if p==0:
        r[n][0]=x0+1
    elif p==1:
        r[n][1]=y0+1
    elif p==2:
        r[n][0]=x0-1
    else:
        r[n][1]=y0-1
    if r[n][0]==L or r[n][1]==L or r[n][0]==-L or r[n][1]==-L:
        r[n][0]=x0
        r[n][1]=y0
    if k%1000==0:
        count=np.zeros((8,8))
        pi=[]
        Si=0
        Ni.append(k)
        for d in range(8):
            for s in range(8):
                for q in range(N):
                    if -100+(25*d) <=r[q][0]<= -75+(25*d) and -100+(25*s) <=r[q][1]<= -75+(25*s):
                          count[d][s]=count[d][s]+1
        for h in range(8):
            for b in range(8):
                pi.append(count[h][b]/N)
        for i in range (64):
            if pi[i] != 0:
                Si=Si+(pi[i]*np.log(pi[i]))
        S.append(-Si)
ra=np.array(r)
plt.xlim(-100,100)
plt.ylim(-100,100)
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal',adjustable='box')
plt.plot(ra[:,0],ra[:,1],'o')
plt.show()
```
